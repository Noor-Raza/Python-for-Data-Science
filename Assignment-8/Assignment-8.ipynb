# **Loading the Dataset and importing necessary libraries**

```python
import pandas as pd
import numpy as np

data = {
    'Student Name': ['Amelie', 'Edgar', 'Jordi', 'Marçal', 'Pep', 'Jaume', 'Marco', 
                     'Ludmila', 'Bastian', 'Marc', 'Gerardo', 'Javier', 'Frank', 'Julia'],
    'Course': ['Artificial Intelligence', 'Python for Data Science', 'Cloud Computing', 'Computer Vision', 
               'Artificial Intelligence', 'Python for Data Science', 'Cloud Computing', 'Computer Vision', 
               'Artificial Intelligence', 'Python for Data Science', 'Cloud Computing', 'Computer Vision', 
               'Artificial Intelligence', 'Python for Data Science'],
    'Grade': [85, 92, 78, 88, 91, 84, 73, 95, 90, 80, 85, 87, 89, 92],
    'Gender': ['Female', 'Male', 'Male', 'Male', 'Male', 'Male', 'Male', 'Female', 'Male', 'Male', 'Male', 'Male', 'Male', 'Female'],
    'Study Time': [15, 20, 10, 8, 18, 14, 11, 16, 19, 13, 16, 12, 17, 15],
    'Age': [22, 21, 23, 22, 20, 24, 22, 21, 23, 21, 22, 22, 21, 20]
}

df = pd.DataFrame(data)
df.head(5)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

---

# **Exercises**

### Exercise 1: Create a *lineplot* showing how **Study Time** varies by **Student Name**. Which student has the highest study time?

**Description:** We are plotting a line to show how much time each student spends studying.

```python
# Creating a lineplot to show how study time changes for each student

plt.figure(figsize=(10, 6))  # Setting the figure size

sns.lineplot(data=df, x='Student Name', y='Study Time', marker='o')  # Plotting the lineplot

plt.title('Study Time by Student Name')  # Adding a title to the chart

plt.xlabel('Student Name')  # Labeling the x-axis

plt.ylabel('Study Time (Hours)')  # Labeling the y-axis

plt.tight_layout()  # Adjusting the layout so that nothing overlaps

plt.show()  # Displaying the plot
```

---

### Exercise 2: Plot a histogram (*histplot*) of **Grade** and determine which grade range has the highest frequency of students.

**Description:** We are creating a bar chart to see how many students fall into each grade range.

```python
# Creating a histogram to show the distribution of grades

plt.figure(figsize=(10, 6))  # Setting the figure size

sns.histplot(data=df, x='Grade', bins=10, kde=True)  # Creating the histogram with a line (kde=True) showing the density

plt.title('Grade Distribution')  # Adding a title to the chart

plt.xlabel('Grade')  # Labeling the x-axis

plt.ylabel('Frequency')  # Labeling the y-axis

plt.tight_layout()  # Adjusting the layout so that nothing overlaps

plt.show()  # Displaying the plot
```

---

### Exercise 3: Create an ECDF plot (*ecdfplot*) for **Grade**. What is the percentage of students scoring less than 85?

**Description:** We are plotting cumulative percentages of grades and calculating the percentage of students with grades below 85.

```python
# Creating an ECDF plot to show cumulative percentages for grades

plt.figure(figsize=(10, 6))  # Setting the figure size

sns.ecdfplot(data=df, x='Grade')  # Creating the ECDF plot for grades

plt.title('ECDF of Grades')  # Adding a title to the chart

plt.xlabel('Grade')  # Labeling the x-axis

plt.ylabel('ECDF')  # Labeling the y-axis

plt.tight_layout()  # Adjusting the layout

plt.show()  # Displaying the plot

# Calculating the percentage of students scoring less than 85

percentage_less_than_85 = (df['Grade'] < 85).mean() * 100  # Finding how many grades are less than 85 and convert to percentage

print(f"Percentage of students scoring less than 85: {percentage_less_than_85:.2f}%")  # Printing the result
```

### Exercise 4: Create a *stripplot* showing **Grade** distribution for each **Course**. Which course has the most spread in grades?

**Description:** We are plotting individual grades for each course to see their spread.

```python
# Creating a stripplot to show how grades are distributed for each course

plt.figure(figsize=(10, 6))  # Setting the figure size

sns.stripplot(data=df, x='Course', y='Grade', jitter=True)  # Creating the stripplot, adding some jitter (spread) for clarity

plt.title('Grade Distribution by Course')  # Adding a title to the chart

plt.xlabel('Course')  # Labeling the x-axis

plt.ylabel('Grade')  # Labeling the y-axis

plt.tight_layout()  # Adjusting the layout

plt.show()  # Displaying the plot
```

---

### Exercise 5: Create a *swarmplot* to show the relationship between Gender and **Study Time**. Which gender has a higher average study time?

**Description:** We are showing the distribution of study times for males and females.

```python
# Creating a swarmplot to show study time distribution for each gender

plt.figure(figsize=(10, 6))  # Setting the figure size

sns.swarmplot(data=df, x='Gender', y='Study Time')  # Creating the swarmplot

plt.title('Study Time by Gender')  # Adding a title to the chart

plt.xlabel('Gender')  # Labeling the x-axis

plt.ylabel('Study Time (Hours)')  # Labeling the y-axis

plt.tight_layout()  # Adjusting the layout

plt.show()  # Displaying the plot
```

----

### Exercise 6: Plot a *pointplot* to show the average **Grade** for each Course. Which course has the highest average grade?

**Description:** We are plotting the average grade for each course to compare performance.

```python
# Creating a pointplot to show the average grade for each course

plt.figure(figsize=(10, 6))  # Setting the figure size

sns.pointplot(data=df, x='Course', y='Grade', estimator='mean')  # Creating the pointplot with averages (mean) for grades

plt.title('Average Grade by Course')  # Adding a title to the chart

plt.xlabel('Course')  # Labeling the x-axis

plt.ylabel('Average Grade')  # Labeling the y-axis

plt.tight_layout()  # Adjusting the layout

plt.show()  # Displaying the plot
```
